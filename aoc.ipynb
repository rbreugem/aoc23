```python
def is_symbol(char:str):
    return (not char.isdigit()) & (char!=".")
```

```python
def find_indices_of_numbers(input_string):
    # Use regular expression to find all matches of numbers in the string
    matches = re.finditer(r'\d+', input_string)

    # Extract and return the starting indices of the matches
    indices = [(match.group(0), match.start(), match.end()-1) for match in matches]

    return indices
```

```python
def number_near_symbol(row_number, start_pos, end_pos, lines):
    
    near_symbol = False
    for relative_row_number in [-1,0,1]:
        for col_number in np.arange(start_pos-1, end_pos+1):
            target_row_number = row_number + relative_row_number
            if target_row_number >= 0 and target_row_number < len(lines):
                if col_number>=0 and col_number < len(lines[target_row_number]):
                    char = lines[target_row_number][col_number]
                    if is_symbol(char):
                        near_symbol = True

    return near_symbol              
            
```

```python
import re

def extract_numbers_with_adjacent_symbol(lines):
    
    som = 0 
    nmbs = {}
    nmbs['numbers']=[]
    nmbs['to_add']=[]
    
    for row_number, line in enumerate(lines):
        indices = find_indices_of_numbers(line)
        for i in indices:
    
            n = i[0]
            nmbs['numbers'].append(n)
            start_pos = i[1]
            end_pos = i[2]
            ns = number_near_symbol(row_number, start_pos=start_pos, end_pos=end_pos, lines=lines)
            
            # print('number:', n, 'near:', ns, 'row:', row_number)
            
            if ns:
                nmbs['to_add'].append(n)
                som = som + int(n)
    return som, nmbs

```

```python
def find_indices_of_gears(lines):
    result = []
    for row_number, line in enumerate(lines):
        # Use regular expression to find all matches of numbers in the string
        matches = re.finditer('\\*', line)

        # Extract and return the starting indices of the matches
        indices = [(match.group(0), match.start(), match.end()-1) for match in matches]
        
        for idx in indices:
            result.append((row_number,idx[1])) 
    return result
```

```python
def get_gear_ratio(gear_pos, lines):
    numbers = []
    numbers_matched = [[],[],[]]
    
    for row_idx, r in enumerate([-1,0,1]):

        rn = gear_pos[0]+r
        
        
        if rn<0 or rn>len(lines):
            continue
        else:
            
            line = lines[rn]
            matches = find_indices_of_numbers(line)
            num_matched = [False for _ in matches]
            
            for c in [-1,0,1]:
                cn = gear_pos[1]+c
                
                if (cn>=0) and (cn<140):
                    for idx, match in enumerate(matches):
                        if (cn>=match[1]) and (cn<=match[2]):
                            num_matched[idx]=True
                    
        for i,nm in enumerate(num_matched):
            if nm:
                numbers.append(int(matches[i][0]))
    return numbers
                        
```

```python
def som_all(gears, lines):
    som = 0
    result = []
    for gear in gears:
        numbers = get_gear_ratio(gear, lines)
        if len(numbers)==2:
            ratio = np.prod(numbers)
            som = som + ratio
        
        print('gear:', str(gear), '  numbers:', str(numbers),'  som:', som)

    return som
```
